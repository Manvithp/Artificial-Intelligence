```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# Load the breast cancer dataset from sklearn
data = load_breast_cancer()

# Create a pandas DataFrame using the feature data and column names from the dataset
dataset = pd.DataFrame(data=data['data'], columns = data['feature_names'])

# Display the DataFrame
dataset
```

```python
x = dataset.copy()
y = data['target']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=69)
```

```python
from sklearn.tree import DecisionTreeClassifier

# dtc = DecisionTreeClassifier()
# ccp_alpha=0.01 - used to prune the tree i.e if it is overfitting then this solves the issue.
dtc = DecisionTreeClassifier(ccp_alpha=0.01)
dtc.fit(x_train, y_train)
```

```python
y_train_pred = dtc.predict(x_train)
y_test_pred = dtc.predict(x_test)
```

```python
# Get all the parameters and their current values from the decision tree classifier
# This returns a dictionary with parameter names as keys and their values as values
# Useful for understanding the current configuration of the model
dtc.get_params()
```

```python
# Use the trained Decision Tree Classifier to predict class probabilities
# So basically it tells the probability of being true or false
# change max_depth to 4 in the model to see changes
dtc.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score

y_train_pred_acc = accuracy_score(y_train, y_train_pred) * 100
y_test_pred_acc = accuracy_score(y_test, y_test_pred) * 100
```

```python
y_test_pred_acc
```

```python
from sklearn.metrics import confusion_matrix

y_train_confMat = confusion_matrix(y_train, y_train_pred, labels=[0,1])
y_test_confMat = confusion_matrix(y_test, y_test_pred, labels=[0,1])
```

```python
from sklearn.metrics import precision_score

y_train_precision_score = precision_score(y_train, y_train_pred)
y_test_precision_score = precision_score(y_test, y_test_pred)
```

```python
from sklearn.metrics import recall_score

y_train_recall = recall_score(y_train, y_train_pred)
y_test_recall = recall_score(y_test, y_test_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_train, y_train_pred, target_names=['malignant', 'benign']))
print(classification_report(y_test, y_test_pred, target_names=['malignant', 'benign']))
```

```python
# Access the feature_importances_ attribute of the decision tree classifier (dtc)
# This returns an array with the importance of each feature in the model's decision making
# Higher values indicate features that have more influence on the model's predictions

dtc.feature_importances_
```

```python
# Create the feature importance DataFrame
fi = pd.DataFrame([dtc.feature_importances_], columns=x.columns).transpose()

# Rename the column - this is the correct syntax
fi.columns = ['Importance']  # Use .columns (plural) instead of .column and assign directly
fi.sort_values(by='Importance', ascending=False, inplace=True)
fi
```

```python
fi.head(10).plot(kind='bar')
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtc, 
                   feature_names=x.columns,  
                   class_names={0:'Malignant', 1:'Benign'},
                   filled=True,
                  fontsize=12)
```
